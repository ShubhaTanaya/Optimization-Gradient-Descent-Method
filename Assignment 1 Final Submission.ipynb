```python
#importing the necessary library
import matplotlib.pyplot as plt
import numpy as np

# enter the function in terms of x, y 
# Gradient Descent estimates the lowest outcome of the function numerically 
def function(x, y):
    # you can edit the function here
    return 5*x**2 + y**2 -30*x - 8*y + 34 +4*x*y

# Calculate the partial derivative with respect to x, using definition of derivative and epsilon is taken as 1e-6
def partial_derivative_x(x, y, epsilon=1e-6):
    deriv = (function(x + epsilon, y) - function(x, y)) / epsilon
    return deriv

# Calculate the partial derivative with respect to y
def partial_derivative_y(x, y, epsilon=1e-6):
    deriv = (function(x, y + epsilon) - function(x, y)) / epsilon
    return deriv

# Gradient Descent function: it takes input of learning rate, number of iterations, and initial coordinates(should be in the domain of function) 
def gdm(learning_rate, iterations, initial_x, initial_y):  
    x = initial_x
    y = initial_y
    
    # creating an array that stores the values of f(x, y)
    f_values = []
    x_values=[]
    y_values=[]
    
    # for each iteration we are defining a new x and y by subtracting a particular step size from the previous x and y
    # the step size is along the direction of the derivative at that point multiplied by the learning rate
    for i in range(iterations):
        
        deri_x = partial_derivative_x(x, y)
        deri_y = partial_derivative_y(x, y)
        
        x = x - learning_rate * deri_x
        y = y - learning_rate * deri_y

        # adding various values of f(x, y) to the f_values array
        f_values.append(function(x, y))
        x_values.append(x)
        y_values.append(y)
        
        # we are also printing the x, y after each iteration to see how the step size reduces as we proceed towards the minima
        print(f"Iteration {i + 1}: x = {x}, y = {y}, f(x, y) = {function(x, y)}")

    # Plotting f(x, y), x, and y vs Iterations using subplot to create a figure with three subplots arranged vertically.
    
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(5, 10))
    #creating a subplot with 3 rows and 1 column, with the specified width and height
    ax1.plot(range(1, iterations + 1), f_values, label='f(x, y)')
    ax1.set_title('f(x, y) vs Iterations')
    ax1.set_ylabel('f(x, y)')
    

    ax2.plot(range(1, iterations + 1), x_values, label='x')
    ax2.set_title('x vs Iterations')
    ax2.set_ylabel('x')
    

    ax3.plot(range(1, iterations + 1), y_values, label='y')
    ax3.set_title('y vs Iterations')
    ax3.set_xlabel('Iterations')
    ax3.set_ylabel('y')
   
    return x, y

# We get the input parameters from the user
learning_rate = float(input("Enter the learning rate: "))
iterations = int(input("Enter the number of iterations: "))
initial_x = float(input("Enter the initial x point: "))
initial_y = float(input("Enter the initial y point: "))

# Running Gradient Descent
final_x, final_y = gdm(learning_rate, iterations, initial_x, initial_y)

# Printing the results for the optimal point and optimal value of the function
print(f"\n Optimal point: x = {final_x}, y = {final_y}")
print(f"Optimal value is: {function(final_x, final_y)}")
```
